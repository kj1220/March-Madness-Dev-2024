```python
import pandas as pd

# Build lake with team info
teams = pd.read_csv('pipe_current_teams.csv')
print('Beginning Teams: ', teams.shape)

# Join BPI
bpi = pd.read_csv('raw_current_bpi.csv')
df = teams.merge(bpi, left_on='bpi', right_on='TEAM', how='inner')
print('ESPN BPI   Join: ', df.shape)

# Join Bart Torvik
torvik = pd.read_csv('raw_current_torvik.csv')
torvik.columns = ['Team_x',
                  'AdjOE_x',
                  'AdjDE_x',
                  'Barthag',
                  'Record', 
                  'Wins',
                  'Games',
                  'EFG%',
                  'EFGD%',
                  'FTR',
                  'FTRD',
                  'TO%',
                  'TOD%',
                  'ORB%',
                  'DRB%',
                  'AdjT',
                  '2P%',
                  '2P%D',
                  '3P%',
                  '3P%D',
                  'na_1',
                  'na_2',
                  'na_3',
                  'na_4',
                  'na_5',
                  'na_6',
                  'na_7',
                  'empty_1',
                  'empty_2',
                  'empty_3',
                  'Year',
                  'empty_4',
                  'empty_5',
                  'empty_6',
                  'WAB',
                  'FT%',
                  'FT%D'
                 ]
df = df.merge(torvik, left_on='torvik', right_on='Team_x', how='inner')
print('BartTorvik Join: ', df.shape)

# Join Ken Pom Summary
kenpom_sum = pd.read_csv('raw_current_kenpom_summary.csv')
df = df.merge(kenpom_sum, left_on='kenpom', right_on='TeamName', how='inner')
print('KenPom Sum Join: ', df.shape)

# Join Ken Pom Point Distribution
kenpom_pts = pd.read_csv('raw_current_kenpom_pointdist.csv')
df = df.merge(kenpom_pts, left_on='kenpom', right_on='TeamName', how='inner')
print('KenPom Pts Join: ', df.shape)

# Join Ken Pom Roster
kenpom_htex = pd.read_csv('raw_current_kenpom_heightex.csv')
df = df.merge(kenpom_htex, left_on='kenpom', right_on='TeamName', how='inner')
print('KenPom HEx Join: ', df.shape)

# Join Miya Ratings
miya = pd.read_csv('raw_current_miya_ratings.csv')
df = df.merge(miya, left_on='miya', right_on='team', how='inner')
print('Miya Join: ', df.shape)

# Filter to only tournament teams
tourney_teams = pd.read_csv('pipe_tourney2024_teams.csv')
df = tourney_teams.merge(df, on='team_year', how='inner')
print('Tourney teams: ', df.shape)
```
