# I. Raw Data

```python
import pandas as pd

# Build lake with team info
teams = pd.read_csv('pipe_historical_teams.csv')
print('Beginning Teams: ', teams.shape)
# Join BPI
bpi = pd.read_csv('raw_historical_bpi.csv')
df = teams.merge(bpi, on='team_year', how='inner')
print('ESPN BPI   Join: ', df.shape)
# Join Bart Torvik
torvik = pd.read_csv('raw_historical_torvik.csv')
df = df.merge(torvik, on='team_year', how='inner')
print('BartTorvik Join: ', df.shape)
# Join Ken Pom Summary
kenpom_sum = pd.read_csv('raw_historical_kenpom_summary.csv')
df = df.merge(kenpom_sum, on='team_year', how='inner')
print('KenPom Sum Join: ', df.shape)
# Join Ken Pom Point Distribution
kenpom_pts = pd.read_csv('raw_historical_kenpom_pointdist.csv')
df = df.merge(kenpom_pts, on='team_year', how='inner')
print('KenPom Pts Join: ', df.shape)
# Join Ken Pom Roster
kenpom_htex = pd.read_csv('raw_historical_kenpom_heightex.csv')
df = df.merge(kenpom_htex, on='team_year', how='inner')
print('KenPom HEx Join: ', df.shape)
```
