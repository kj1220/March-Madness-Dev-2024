```python
import numpy as np
import pandas as pd

games = pd.read_csv('pipe_historical_games_weekend2.csv')
print(games.shape)
ratings = pd.read_csv('ware_historical_team_ratings.csv')
print(ratings.shape)
four_factors = pd.read_csv('ware_historical_four_factors.csv')
print(four_factors.shape)
stats = pd.read_csv('ware_historical_team_stats.csv')
print(stats.shape)
style = pd.read_csv('ware_historical_team_style.csv')
print(style.shape)
tiers = pd.read_csv('mart_page1_teams.csv')
print(tiers.shape)
```

```python
df = ratings.merge(four_factors, on='team_year', how='inner')
print(df.shape)
df = df.merge(stats, on='team_year', how='inner')
print(df.shape)
df = df.merge(style, on='team_year', how='inner')
print(df.shape)
df = df.merge(tiers, on='team_year', how='inner')
print(df.shape)
```

# III. Merge Stats for Low Bracket

```python
df = df.rename(columns={
 'team_year': 'low_team',
 'ovr_rating_bpi': 'low_ovr_rating_bpi',
 'ovr_rating_kenpom': 'low_ovr_rating_kenpom',
 'ovr_rating_torvik': 'low_ovr_rating_torvik',
 'wab': 'low_wab',
 'off_rating_bpi': 'low_off_rating_bpi',
 'off_rating_kenpom': 'low_off_rating_kenpom',
 'off_rating_torvik': 'low_off_rating_torvik',
 'def_rating_bpi': 'low_def_rating_bpi',
 'def_rating_kenpom': 'low_def_rating_kenpom',
 'def_rating_torvik': 'low_def_rating_torvik',
 'efg%_off': 'low_efg%_off',
 'ftr_off': 'low_ftr_off',
 'reb%_off': 'low_reb%_off',
 'to%_off': 'low_to%_off',
 'efg%_def': 'low_efg%_def',
 'ftr_def': 'low_ftr_def',
 'reb%_def': 'low_reb%_def',
 'to%_def': 'low_to%_def',
 '2p%_off': 'low_2p%_off',
 '3p%_off': 'low_3p%_off',
 'eff_off': 'low_eff_off',
 'ft%_off': 'low_ft%_off',
 '2p%_def': 'low_2p%_def',
 '3p%_def': 'low_3p%_def',
 'eff_def': 'low_eff_def',
 'ft%_def': 'low_ft%_def',
 'height_eff': 'low_height_eff',
 'size': 'low_size',
 'bench': 'low_bench',
 'adj_tempo': 'low_adj_tempo',
 'tier': 'low_tier'
})
```

# IV. Create Matchup Stats

```python
# Scale Defensive Efficiency Statistics
df['high_def_rating_kenpom_scaled'] = (1/(df[['high_def_rating_kenpom']]/100))*100
df['high_def_rating_torvik_scaled'] = (1/(df[['high_def_rating_torvik']]/100))*100
df['low_def_rating_kenpom_scaled'] = (1/(df[['low_def_rating_kenpom']]/100))*100
df['low_def_rating_torvik_scaled'] = (1/(df[['low_def_rating_torvik']]/100))*100
```

```python
matchup_df = df[['game_id',
                 'id',
                 'year',
                 'region',
                 'round',
                 'high_bracket_seed',
                 'high_bracket_team',
                 'low_bracket_seed',
                 'low_bracket_team',
                 'high_bracket_score',
                 'low_bracket_score',
                 'point_diff',
                 'point_total',
                 'win']]
```

```python
# Overall ratings
matchup_df['ovr_rating_bpi'] = df['high_ovr_rating_bpi'] - df['low_ovr_rating_bpi']
matchup_df['ovr_rating_kenpom'] = df['high_ovr_rating_kenpom'] - df['low_ovr_rating_kenpom']
matchup_df['ovr_rating_torvik'] = df['high_ovr_rating_torvik'] - df['low_ovr_rating_torvik']
matchup_df['wab'] = df['high_wab'] - df['low_wab']
# Offensive ratings
matchup_df['off_rating_bpi'] = df['high_off_rating_bpi'] - df['low_def_rating_bpi']
matchup_df['off_rating_kenpom'] = df['high_off_rating_kenpom'] - df['low_def_rating_kenpom_scaled']
matchup_df['off_rating_torvik'] = df['high_off_rating_torvik'] - df['low_def_rating_torvik_scaled']
# Defensive ratings
matchup_df['def_rating_bpi'] = df['high_def_rating_bpi'] - df['low_off_rating_bpi']
matchup_df['def_rating_kenpom'] = df['high_def_rating_kenpom_scaled'] - df['low_off_rating_kenpom']
matchup_df['def_rating_torvik'] = df['high_def_rating_torvik_scaled'] - df['low_off_rating_torvik']
# Four factors offense
matchup_df['efg%_off'] = df['high_efg%_off'] - df['low_efg%_def']
matchup_df['ftr_off'] = df['high_ftr_off'] - df['low_ftr_def']
matchup_df['reb%_off'] = df['high_reb%_off'] - df['low_reb%_def']
matchup_df['to%_off'] = df['high_to%_off'] - df['low_to%_def']
# Four factors defense
matchup_df['efg%_def'] = df['high_efg%_def'] - df['low_efg%_off']
matchup_df['ftr_def'] = df['high_ftr_def'] - df['low_ftr_off']
matchup_df['reb%_def'] = df['high_reb%_def'] - df['low_reb%_off']
matchup_df['to%_def'] = df['high_to%_def'] - df['low_to%_off']
# Shooting offense
matchup_df['2p%_off'] = df['high_2p%_off'] - df['low_2p%_def']
matchup_df['3p%_off'] = df['high_3p%_off'] - df['low_3p%_def']
matchup_df['eff_off'] = df['high_eff_off'] - df['low_eff_def']
matchup_df['ft%_off'] = df['high_ft%_off'] - df['low_ft%_def']
# Shooting defense
matchup_df['2p%_def'] = df['high_2p%_def'] - df['low_2p%_off']
matchup_df['3p%_def'] = df['high_3p%_def'] - df['low_3p%_off']
matchup_df['eff_def'] = df['high_eff_def'] - df['low_eff_off']
matchup_df['ft%_def'] = df['high_ft%_def'] - df['low_ft%_off']
# Miscellaneous
matchup_df['height_eff'] = df['high_height_eff'] - df['low_height_eff']
matchup_df['size'] = df['high_size'] - df['low_size']
matchup_df['bench'] = df['high_bench'] - df['low_bench']
matchup_df['adj_tempo'] = df['high_adj_tempo'] - df['low_adj_tempo']
# S Tiers
matchup_df['s_s'] = df.apply(lambda row: 1 if row['high_tier'] == 'S' and row['low_tier'] == 'S' else 0, axis=1)
matchup_df['s_a'] = df.apply(lambda row: 1 if row['high_tier'] == 'S' and row['low_tier'] == 'A' else 0, axis=1)
matchup_df['s_b'] = df.apply(lambda row: 1 if row['high_tier'] == 'S' and row['low_tier'] == 'B' else 0, axis=1)
matchup_df['s_c'] = df.apply(lambda row: 1 if row['high_tier'] == 'S' and row['low_tier'] == 'C' else 0, axis=1)
matchup_df['s_d'] = df.apply(lambda row: 1 if row['high_tier'] == 'S' and row['low_tier'] == 'D' else 0, axis=1)
matchup_df['s_f'] = df.apply(lambda row: 1 if row['high_tier'] == 'S' and row['low_tier'] == 'F' else 0, axis=1)
# A Tiers
matchup_df['a_s'] = df.apply(lambda row: 1 if row['high_tier'] == 'A' and row['low_tier'] == 'S' else 0, axis=1)
matchup_df['a_a'] = df.apply(lambda row: 1 if row['high_tier'] == 'A' and row['low_tier'] == 'A' else 0, axis=1)
matchup_df['a_b'] = df.apply(lambda row: 1 if row['high_tier'] == 'A' and row['low_tier'] == 'B' else 0, axis=1)
matchup_df['a_c'] = df.apply(lambda row: 1 if row['high_tier'] == 'A' and row['low_tier'] == 'C' else 0, axis=1)
matchup_df['a_d'] = df.apply(lambda row: 1 if row['high_tier'] == 'A' and row['low_tier'] == 'D' else 0, axis=1)
matchup_df['a_f'] = df.apply(lambda row: 1 if row['high_tier'] == 'A' and row['low_tier'] == 'F' else 0, axis=1)
# B Tiers
matchup_df['b_s'] = df.apply(lambda row: 1 if row['high_tier'] == 'B' and row['low_tier'] == 'S' else 0, axis=1)
matchup_df['b_a'] = df.apply(lambda row: 1 if row['high_tier'] == 'B' and row['low_tier'] == 'A' else 0, axis=1)
matchup_df['b_b'] = df.apply(lambda row: 1 if row['high_tier'] == 'B' and row['low_tier'] == 'B' else 0, axis=1)
matchup_df['b_c'] = df.apply(lambda row: 1 if row['high_tier'] == 'B' and row['low_tier'] == 'C' else 0, axis=1)
matchup_df['b_d'] = df.apply(lambda row: 1 if row['high_tier'] == 'B' and row['low_tier'] == 'D' else 0, axis=1)
matchup_df['b_f'] = df.apply(lambda row: 1 if row['high_tier'] == 'B' and row['low_tier'] == 'F' else 0, axis=1)
# C Tiers
matchup_df['c_s'] = df.apply(lambda row: 1 if row['high_tier'] == 'C' and row['low_tier'] == 'S' else 0, axis=1)
matchup_df['c_a'] = df.apply(lambda row: 1 if row['high_tier'] == 'C' and row['low_tier'] == 'A' else 0, axis=1)
matchup_df['c_b'] = df.apply(lambda row: 1 if row['high_tier'] == 'C' and row['low_tier'] == 'B' else 0, axis=1)
matchup_df['c_c'] = df.apply(lambda row: 1 if row['high_tier'] == 'C' and row['low_tier'] == 'C' else 0, axis=1)
matchup_df['c_d'] = df.apply(lambda row: 1 if row['high_tier'] == 'C' and row['low_tier'] == 'D' else 0, axis=1)
matchup_df['c_f'] = df.apply(lambda row: 1 if row['high_tier'] == 'C' and row['low_tier'] == 'F' else 0, axis=1)
# D Tiers
matchup_df['d_s'] = df.apply(lambda row: 1 if row['high_tier'] == 'D' and row['low_tier'] == 'S' else 0, axis=1)
matchup_df['d_a'] = df.apply(lambda row: 1 if row['high_tier'] == 'D' and row['low_tier'] == 'A' else 0, axis=1)
matchup_df['d_b'] = df.apply(lambda row: 1 if row['high_tier'] == 'D' and row['low_tier'] == 'B' else 0, axis=1)
matchup_df['d_c'] = df.apply(lambda row: 1 if row['high_tier'] == 'D' and row['low_tier'] == 'C' else 0, axis=1)
matchup_df['d_d'] = df.apply(lambda row: 1 if row['high_tier'] == 'D' and row['low_tier'] == 'D' else 0, axis=1)
matchup_df['d_f'] = df.apply(lambda row: 1 if row['high_tier'] == 'D' and row['low_tier'] == 'F' else 0, axis=1)
# F Tiers
matchup_df['f_s'] = df.apply(lambda row: 1 if row['high_tier'] == 'F' and row['low_tier'] == 'S' else 0, axis=1)
matchup_df['f_a'] = df.apply(lambda row: 1 if row['high_tier'] == 'F' and row['low_tier'] == 'A' else 0, axis=1)
matchup_df['f_b'] = df.apply(lambda row: 1 if row['high_tier'] == 'F' and row['low_tier'] == 'B' else 0, axis=1)
matchup_df['f_c'] = df.apply(lambda row: 1 if row['high_tier'] == 'F' and row['low_tier'] == 'C' else 0, axis=1)
matchup_df['f_d'] = df.apply(lambda row: 1 if row['high_tier'] == 'F' and row['low_tier'] == 'D' else 0, axis=1)
matchup_df['f_f'] = df.apply(lambda row: 1 if row['high_tier'] == 'F' and row['low_tier'] == 'F' else 0, axis=1)
```
